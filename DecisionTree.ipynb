### Import Library

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeClassifier ,  plot_tree #DecisionTreeClassifier  (model)
```

```python
df = pd.read_csv ('Iris.csv')  #y indepandent x dependent
df
```

```python
df = df.drop('Id',axis=1)
df 
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Species'].unique()
```

```python
df.replace({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2}, inplace=True) 
df
```

```python
x = df.drop('Species',axis=1) #
y = df['Species']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4)
```

```python
model = DecisionTreeClassifier(max_depth=4) #max dari kedalaman treenya
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test) 
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)*100
r2
```

```python
plt.figure(figsize=(8,7))

plot_tree(model,filled=True,class_names=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
        feature_names=['SepalLenghtCm', 'SepalWidhtCm', 'PetalLeghtCm', 'PetalWidhtCm'])
plt.show()
```
